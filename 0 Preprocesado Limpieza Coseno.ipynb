```python
path = "./versions/version_2"

# Lista de carpetas en el dataset original
folder_group = [name for name in os.listdir(path) if os.path.isdir(os.path.join(path, name))]
folder_group
```

```python
df = pd.DataFrame(columns=['Name1','Code1','Name2', 'Code2'])

for folder in folder_group:

    dataset1 = []
    dataset2 = []
    # aux = []

    folder_path = os.path.join(path, folder)
    problem_list = os.listdir(folder_path)
    # print(problem_list[0])
    # print(problem_list[1])

    # Eliminar extensión ".java" del nombre del archivo
    problem_list = [os.path.splitext(file)[0] for file in problem_list]

    file1 = open(os.path.join(folder_path, problem_list[0] + '.java'), 'r')
    dataset1.append(file1.read())
    # aux.append(file1.read().split('/n'))
    # print(file1.read())
    file1.close()
    
    d1 = ' '.join(dataset1)

    file2 = open(os.path.join(folder_path, problem_list[1] + '.java'), 'r')
    dataset2.append(file2.read())
    # aux.append(file2.read())
    # df['Code2'] = dataset2
    file2.close()

    d2 = ' '.join(dataset2)

    # X = vectorizer.fit_transform(aux)
    # similarity_matrix = cosine_similarity(X,X)

    # TD-IDF Matrix
    df.loc[len(df.index)] = [problem_list[0], d1, problem_list[1], d2]
    # df.loc[len(df.index)] = [d1, d2, similarity_matrix[0][1]]

```

```python
def calculate_similarity(code1, code2):
    vectorizer = TfidfVectorizer()

    # Crear la matriz de características TF-IDF
    tfidf_matrix = vectorizer.fit_transform([code1, code2])
    
    # Calcular la similitud entre los dos códigos
    similarity = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]

    return similarity
```

```python
df['Similitud'] = df.apply(lambda row: calculate_similarity(row['Code1'], row['Code2']), axis=1)
```
